```python
plt.figure(figsize=(8,4))
sns.barplot(x=sui_cide['sex'], y=sui_cide['suicides_no'], hue=sui_cide['age'])
plt.title('Suicide Rate Sex and Age')
plt.show()
```

The diagram above shows that the suicide rate of males in 1999 was higher than that of females, and this rate is between the age of 35 and 54
